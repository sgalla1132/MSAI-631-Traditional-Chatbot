```python
def chatbot_response(user_input):
    user_input = user_input.lower().strip()

    if user_input in ["hello", "hi", "hey"]:
        return "Hello! I am Campus Helper. Type 'help' to see what I can do."

    elif user_input == "help":
        return "I can help with: assignments, deadlines, instructor info, course resources, and exit."

    elif "assignment" in user_input:
        return "This project requires creating a simple traditional chatbot and documenting the development process."

    elif "deadline" in user_input:
        return "Please check the course shell for the official deadline. Most assignments are due by 11:59 PM."

    elif "instructor" in user_input:
        return "Your instructor is Professor Ying Lin. Please use the course email or announcement for official contact."

    elif "resource" in user_input or "github" in user_input:
        return "You should include your GitHub repository URL and submit a ZIP archive of your project."

    elif user_input in ["bye", "exit", "quit"]:
        return "Goodbye! Good luck with your assignment."

    else:
        return "Sorry, I did not understand that. Type 'help' to see what I can answer."


print("Campus Helper Chatbot")
print("Type 'bye', 'exit', or 'quit' to stop.")

while True:
    user_input = input("You: ")
    response = chatbot_response(user_input)
    print("Bot:", response)

    if user_input.lower().strip() in ["bye", "exit", "quit"]:
        break
```
